```julia
m = 50
rng = MersenneTwister(1234)  
u = rand(rng, [-1, 1], m)
c = [1, 0.7, -0.3]
h = [0.9, -0.5, 0.5, -0.4, 0.3, -0.3, 0.2, -0.1]

y = conv(c, u)
z = conv(h, y)


plot(1:length(u), u, label= "u", title= "Signal u", xlabel= "Sample Index", ylabel= "Amplitude")
plot(1:length(y), y, label= "y = c ∗ u", title= "Convolution with c", xlabel= "Sample Index", ylabel= "Amplitude")
plot(1:length(z), z, label= "z = h ∗ y", title= "Equalized Signal z", xlabel= "Sample Index", ylabel= "Amplitude")

# Show the plots
display(plot(1:length(u), u, label= "u", title= "Signal u", xlabel= "Sample Index", ylabel= "Amplitude"))
display(plot(1:length(y), y, label= "y = c ∗ u", title= "Convolution y = c ∗ u", xlabel= "Sample Index", ylabel= "Amplitude"))
display(plot(1:length(z), z, label= "z = h ∗ y", title= "Equalized Signal z = h ∗ y", xlabel= "Sample Index", ylabel= "Amplitude"))
```
